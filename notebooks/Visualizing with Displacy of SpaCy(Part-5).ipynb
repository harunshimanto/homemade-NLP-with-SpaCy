* For Jupyter notebook

```python
import spacy 
nlp=spacy.load('en')
```

```python
from spacy import displacy
```

```python
docx=nlp("The rat the cat the dog chased killed ate the malt ")
```

```python
displacy.render(docx,style='dep',jupyter=True)
```

```python
options={'compact' : True,'bg': 'red',
                'color': '#fff','font':'Sans Serif' }
```

```python
displacy.render(docx,style='dep',options=options)
```
